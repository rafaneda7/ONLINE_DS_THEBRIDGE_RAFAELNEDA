```python
def listsum(numbers):
    if not numbers:
        return 0
    else:
        (f,rest) = numbers
        return f + listsum(rest)
    
mylist = (1,(2,(3, None)))
total = listsum(mylist)

print(total)

# ¿Cómo sé que en la definicion de esta funcion solo van a tomar en cuenta los int y float y si pongo algun str, bool o algun otro elemento que no pertenezcan a estas clases me vaya a dar error?
# Segun entiendo si el parametro no es un numero me deberia dar 0, pero si coloco un str me da error quisiera saber porque pasa esto? Es por el (f, rest) y por eso se deberia colocar
# Siempre una lista o una tupla? 


# Explicar el try/except y un ejemplo de su aplicacion, por ejemplo, una division por cero que deberia dar error 

def division(a,b):

    try:
        a_int = int(a)
        b_int = int(b)
        resultado = a_int / b_int
        return resultado
    except(ZeroDivisionError):
        return "No se puede dividir entre cero"
    except(TypeError):
        print("Uno de los parametros no es un numero")
    
division(10,5)
division(10,0)
division("w",10)
```

```python
def division2(a, b):
    a_int = int(a)
    b_int = int(b)
    try:
        resultado = a_int / b_int
        print(f"El resultado de {a} dividido por {b} es {resultado}")
    except:
        print("Error: No se puede dividir por cero.")
division(10, 2)  # Esto funcionará y mostrará el resultado.
division(10, 0)  # Esto lanzará una excepción y mostrará el mensaje de error.
division("w", 74)
```

```python
class Perro:
    patas = 4
    orejas = 2
    ojos = 2

    def __init__(self, raza, color_pelo = "Marron", dueño = "no tiene dueño"):
        self.raza = raza
        self.color_pelo = color_pelo
        self.dueño = dueño
        self.velocidad = 0
    def parar(self):
        self.velocidad = 0
        return self.velocidad
    def andar(self, aumento_velocidad):
        self.velocidad += aumento_velocidad
        return self.velocidad
    def ladrar(self, sonido):
        return "GUAU!" + " " + sonido
    def __str__ (self):
        return f"La raza de mi perro es: {self.raza}, El color de su pelo es: {self.color_pelo}, Este perro {self.dueño}"
    
# ¿Por qué si no coloco esta def de str no me imprime bien los parametros de la clase sino <__main__.Perro object at 0x000001AE30C99790>?

mi_perro = Perro(raza = "puddle")
print(mi_perro)

velocidad_de_mi_perro = mi_perro.andar(5)
print("La velocidad de mi perro al andar es", velocidad_de_mi_perro)

print("La velocidad de mi perro al detenerse es", mi_perro.parar())

print("Mi perro hace", mi_perro.ladrar("Woof Woof"))
```

```python
def comparacion_dos_numeros(num1, num2):
    if num1 == num2:
        print("Los números son iguales.")
    elif num1 > num2:
        print("El primer número es mayor que el segundo.")
    else:
        print("El segundo número es mayor que el primero.")

comparacion_dos_numeros(10,20)
# ¿Cuál es la diferencia si en este ejercicio hubiera colocado un return en vez de un print?
```
